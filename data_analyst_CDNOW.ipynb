```python
rfm = df.pivot_table(index = 'id',
                     values = ['date','quantity','amount'],
                     aggfunc = {'date':'max',
                                'quantity':'count',
                                'amount':'sum'})
rfm['R']=(rfm['date'].max() - rfm['date']) / timedelta(1)
rfm.rename(columns = {'amount':'M', 'quantity':'F'},inplace=True)
rfm=rfm[['R','F','M']]
rfm.head()
```

RFM
,10RFM

```python
def level_label(data):
    level = data.apply(lambda x:'1' if x>0 else '0')
    label = level['R']+level['F']+level['M']
    d = {
        
        '011':"",
        '111':"",
        '001':"",
        '101':"",
        '010':"",
        '000':"",
        '110':"",
        '100':""
    }
    result = d[label]
    return result

rfm['label']=rfm.apply(lambda x:x-x.mean()).apply(level_label,axis=1)
print(rfm.groupby('label').sum())
```

```python
rfm_pie = rfm.groupby('label').size()
pylab.rcParams['figure.figsize']=(7,7)
plt.title("")
plt.pie(rfm_pie,labels=rfm_pie.index,autopct = '%2.1f%%',startangle = 90)
#plt.legend(fancybox=True, loc='best')
plt.axis('equal')     # plot pyplot as circle
plt.tight_layout()  #tight_layout
```

```python
def avtive_status(data):
    status=[]
    for i in range(18): #18
        #
        if data[i]>0:  
            #
            if len(status)==0:
                status.append('new')  #    
            else:
                #
                if status[i-1]=='unreg':
                    status.append('new') #
                #unactive
                elif status[i-1]=='unactive':
                    status.append('return') #
                #newactivereturn
                else:
                    status.append('active') #
        #
        else:
            if len(status)==0:
                status.append('unreg')  #   
            else:
                #
                if status[i-1]=='unreg':
                    status.append('unreg') #
                #newunactiveactivereturn
                else:
                    status.append('unactive') #
                    
    return status

active_count = pivoted_count.apply(lambda x:pd.Series(avtive_status(x),index=pivoted_count.columns),axis=1)
print(active_count.head())
```
